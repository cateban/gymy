```python
import re
import requests
from bs4 import BeautifulSoup
import unidecode
import numpy as np
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
from slugify import slugify
from opencage.geocoder import OpenCageGeocode
import pymongo
from pymongo import MongoClient
from tqdm import tqdm
from tqdm import trange
import folium
from folium import plugins
```

```python
#inicializo conexión con geocoder
key = 'd8314354b31d46a298e0f21ffb920a34'
geocoder = OpenCageGeocode(key)
```

```python
def getting_latlong(x):
    try:
        results = geocoder.geocode(x)
        coordinates=[]
        coordinates.append(results[0]['geometry']['lng'])
        coordinates.append(results[0]['geometry']['lat']) 
        return coordinates
    except:
        return np.nan
```

```python
gymy = pd.read_csv('gymy_V3.csv')
gymy.drop(columns='Unnamed: 0',inplace=True)
```

```python
tqdm.pandas()
gymy['latlong'] = gymy.address.progress_apply(getting_latlong)
```

```python
def long(x):
    try:
        return x[0]
    except:
        return x
    
def lat(x):
    try:
        return x[1]
    except:
        return x

gymy['longitude'] = gymy.latlong.apply(long)
gymy['latitude'] = gymy.latlong.apply(lat)
```

```python
tqdm.pandas()
gymy_names['latlong'] = gymy.names.progress_apply(getting_latlong)
```

```python
gymy = pd.read_csv('gymy_V3 - gymy_V3.csv')
gymy.drop(columns=['id','Unnamed: 9'],inplace=True)
```

```python
gymy.shape
```

```python
gymy['long'] = gymy.link.apply(lambda x : re.search(r'@((\d+\.\d+),(-\d+\.\d+))',x).group(3))
gymy['lat'] = gymy.link.apply(lambda x : re.search(r'@((\d+\.\d+),(-\d+\.\d+))',x).group(2))
```

```python
gymy.head()
```

```python
lista_points = []
for i in tqdm(range(len(gymy))):
    
    lista_points.append(dict({'type':'Point',
                          'coordinates': [gymy.long[i],
                                         gymy.lat[i]]}))
```

```python
#adding geopoints to main DF
gymy['principal'] = pd.Series(lista_points)
```
